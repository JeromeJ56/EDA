```python
# Dropping Top Ten Column
df = df.drop(['Top Ten'], axis=1)
```

```python
# Checking for Null values
df.isnull().sum()
```

```python
#fill with mode (Pack)
df['Style'] = df['Style'].fillna(value='Pack')
df.isna().sum()
```

```python
ramen_style =['Pack','Bowl','Cup','Tray','Box','Bar','Can' ]
```

```python
ramen_style_sizes = df['Style'].value_counts()
ramen_style_sizes = list(ramen_style_sizes)
```

```python
# Donut Plot
fig = plt.figure()
fig.patch.set_facecolor('black')

# Change color of text
plt.rcParams['text.color'] = 'white'
plt.rcParams.update({'font.size': 22})

# Create a circle for the center of the plot
my_circle=plt.Circle( (0,0), 0.7, color='black')

# Pieplot + circle on it
plt.pie(ramen_style_sizes, labels=ramen_style)
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.rcParams["figure.figsize"] = (10,10)
plt.show()
```

```python
# Coutry counts
df['Country'].value_counts()
```

```python
value_counts = df['Country'].value_counts()
df_country = pd.DataFrame(value_counts)
```

```python
top_7_countries = df_country.nlargest(7,"Country")
top_7_countries
```

```python
# Barplot  of Top Ramen Cosuming Countries 
country = list(top_7_countries.index)

values = top_7_countries['Country']
height  = list(values)
y_pos = np.arange(len(country))

# Create bars
plt.bar(y_pos, height,color = ['red','green','blue','yellow','black','purple','brown'])

# Create names on the x-axis
plt.xticks(y_pos, country, color='purple',rotation=90)
plt.yticks(color='orange')

# Show graphic
plt.show()
```
